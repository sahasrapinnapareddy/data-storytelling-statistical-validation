```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Old_Website_Sales': [200, 220, 250, 240, 260, 270],
    'New_Website_Sales': [230, 250, 290, 300, 320, 340]
}

df = pd.DataFrame(data)

df
```

```python
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
```

```python

print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))

plt.plot(df['Month'], df['Old_Website_Sales'], marker='o', label='Old Website')
plt.plot(df['Month'], df['New_Website_Sales'], marker='o', label='New Website')

plt.title("Old vs New Website Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()

plt.show()
```

```python
df.plot(
    x='Month',
    y=['Old_Website_Sales', 'New_Website_Sales'],
    kind='bar',
    figsize=(8,5)
)

plt.title("Monthly Sales Comparison")
plt.ylabel("Sales")

plt.show()
```

```python
old_avg = df['Old_Website_Sales'].mean()
new_avg = df['New_Website_Sales'].mean()

print("Average Old Website Sales:", old_avg)
print("Average New Website Sales:", new_avg)

increase = ((new_avg - old_avg) / old_avg) * 100

print(f"\nPercentage Increase in Sales: {increase:.2f}%")
```

```python
t_stat, p_value = stats.ttest_ind(
    df['Old_Website_Sales'],
    df['New_Website_Sales']
)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Result: Significant difference found.")
    print("The new website increased sales successfully.")
else:
    print("Result: No significant difference found.")
```
